```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
sns.set(style="white")
```

```python
def plot_data(data,labels=None):
    """
    Affiche des donnees 2D
    :param data: matrice des donnees 2d
    :param labels: vecteur des labels (discrets)
    :return:
    """
    cols,marks = ["red", "blue","green", "orange", "black", "cyan"],[".","+","*","o","x","^"]
    if labels is None:
        plt.scatter(data[:,0],data[:,1],marker="x")
        return
    for i,l in enumerate(sorted(list(set(labels.flatten())))):
        plt.scatter(data[labels==l,0],data[labels==l,1],c=cols[i],marker=marks[i])
```

```python
def plot_frontiere(data, f, step=20, alpha_c=1):
    """ Trace un graphe de la frontiere de decision de f
    :param data: donnees
    :param f: fonction de decision
    :param step: pas de la grille
    :return:
    """
    grid,x,y=make_grid(data=data,step=step)
    plt.contourf(x,y,f(grid).reshape(x.shape),256, alpha=alpha_c)
```

```python
def make_grid(data=None,xmin=-5,xmax=5,ymin=-5,ymax=5,step=20):
    """ Cree une grille sous forme de matrice 2d de la liste des points
    :param data: pour calcluler les bornes du graphe
    :param xmin: si pas data, alors bornes du graphe
    :param xmax:
    :param ymin:
    :param ymax:
    :param step: pas de la grille
    :return: une matrice 2d contenant les points de la grille
    """
    if data is not None:
        xmax, xmin, ymax, ymin = np.max(data[:,0]),  np.min(data[:,0]), np.max(data[:,1]), np.min(data[:,1])
    x, y =np.meshgrid(np.arange(xmin,xmax,(xmax-xmin)*1./step), np.arange(ymin,ymax,(ymax-ymin)*1./step))
    grid=np.c_[x.ravel(),y.ravel()]
    return grid, x, y


```

```python
def gen_arti(centerx=1,centery=1,sigma=0.1,nbex=1000,data_type=0,epsilon=0.02):
    """ Generateur de donnees,
        :param centerx: centre des gaussiennes
        :param centery:
        :param sigma: des gaussiennes
        :param nbex: nombre d'exemples
        :param data_type: 0: melange 2 gaussiennes, 1: melange 4 gaussiennes, 2:echequier
        :param epsilon: bruit dans les donnees
        :return: data matrice 2d des donnnes,y etiquette des donnnees
    """
    if data_type==0:
         #melange de 2 gaussiennes
         xpos=np.random.multivariate_normal([centerx,centerx],np.diag([sigma,sigma]),int(nbex//2))
         xneg=np.random.multivariate_normal([-centerx,-centerx],np.diag([sigma,sigma]),int(nbex//2))
         data=np.vstack((xpos,xneg))
         y=np.hstack((np.ones(nbex//2),-np.ones(nbex//2)))
    if data_type==1:
        #melange de 4 gaussiennes
        xpos=np.vstack((np.random.multivariate_normal([centerx,centerx],np.diag([sigma,sigma]),int(nbex//4)),np.random.multivariate_normal([-centerx,-centerx],np.diag([sigma,sigma]),int(nbex/4))))
        xneg=np.vstack((np.random.multivariate_normal([-centerx,centerx],np.diag([sigma,sigma]),int(nbex//4)),np.random.multivariate_normal([centerx,-centerx],np.diag([sigma,sigma]),int(nbex/4))))
        data=np.vstack((xpos,xneg))
        y=np.hstack((np.ones(nbex//2),-np.ones(int(nbex//2))))

    if data_type==2:
        #echiquier
        data=np.reshape(np.random.uniform(-4,4,2*nbex),(nbex,2))
        y=np.ceil(data[:,0])+np.ceil(data[:,1])
        y=2*(y % 2)-1
        
    # un peu de bruit
    data[:,0]+=np.random.normal(0,epsilon,nbex)
    data[:,1]+=np.random.normal(0,epsilon,nbex)
    # on mélange les données
    idx = np.random.permutation((range(y.size)))
    data=data[idx,:]
    y=y[idx]
    return data,y
```

###  Q 1.3 Le modèle du classifieur bayésien naïf fait les hypothèses suivantes 
 • Indépendance des dimensions : $p(x|y) = \prod_{i=1}^{d}p(x_i |y)$
 • Pour chaque dimension i, $p(x_i |y)$ suit une loi normale de paramètre (µi , σi) (ou une bernoulli dans le cas discret binaire). 

 Pour classer un exemple x, la règle de décision consiste à trouver le label qui maximise la probabilité a posteriori:

  $y^{*} = argmax_yp(y|x) = argmax_y \frac {p(x|y)p(y)}{ p(x)} = argmax_yp(y)*\prod_{i=1}^{d} p(x_i|y) $

 L’apprentissage de ce modèle se fait en estimant les paramètres des gaussiennes de chaque dimension pour chaque label par maximum de vraisemblance.

 Pour des raisons de précisions numériques, on utilise souvent le logarithme de la probabilité :
$y^{*} = argmax_ylog(p(y)) + \sum_{ i=1}^{d} log(p(x_i |y)).$
Par ailleurs, les probabilités estimées à 0 sont remplacées par une valeur très faible, 10−10 par exemple (pourquoi ?).

### Q 1.3.1
Implémenter un classifieur bayésien naïf. Vous utiliserez un vecteur µ pour stocker pour chaque classe la moyenne estimée pour chaque dimension,
un vecteur σ pour stocker pour chaque classe la variance pour chaque dimension, et un réel pour stocker la probabilité a priori de chaque classe. Dans les fonctions, n’utiliser qu’une seule boucle for si besoin sur les labels mais surtout pas sur les données : utiliser les fonctions de numpy pour  cela ; utiliser les fonctions np.mean et np.std(ddof=1) pour estimer la moyenne et la variance corrigée des données

```python
def get_mu_sigma(x, y):
    """
    X donnees de dimension d;
    Y labels;
    Retourne deux matrices, mu pour les moyennes,
    std pour l'ecart type
    
    Chaque ligne correspond à une classe et chaque colonne à une dimension
    """
    #Parcours des classes
    
    CLASSES={0:-1, 1:1}
    
    temp_mu = np.mean(x[y==CLASSES[0]], axis=0) 
    temp_sigma = np.std(x[y==CLASSES[0]], axis=0, ddof=1)
    
    temp_mu1= np.mean(x[y==CLASSES[1]], axis=0)
    temp_sigma1 = np.std(x[y==CLASSES[1]], axis=0, ddof=1)
    
    mu = np.vstack([temp_mu, temp_mu1])
    sigma = np.vstack([temp_sigma, temp_sigma1])
    
    # On ajoute du bruit pour les dimensions ou la variance et la moyenne sont nulles
    return mu+10e-10, sigma+10e-10
```

```python
mu, sigma= get_mu_sigma(x,y)
```

```python
def get_log_p_classes(x, y):
    """
    X donnees de dimension d;
    Y labels;
    Retourne  le log de la probabilité empirique de chaque classe 
    """
    CLASSES={0:1, 1:-1}
    return np.log(float(x[y==CLASSES[0]].shape[0])/x.shape[0]), np.log(float(x[y==CLASSES[1]].shape[0])/x.shape[0])
```

```python
def predict_label_bayesian(x, logp1, logp2 , mu, sigma):
    """
    X donnees de dimension d;
    logp1, logp2, les log desprobabilités empiriques de chaque classe 
    mu, sigma matrice des variances et moyenne
    Retourne la prediction bayesienne
    """
    CLASSES = np.array([-1,1])

    critere_2 = logp2 + np.sum(
        -np.log(np.sqrt(2*np.pi)*sigma[1,:]) -0.5*np.power(np.divide(
                    (x-mu[1,:]),sigma[1,:]),2),axis=1)
    
    critere_1 = logp2 + np.sum(
        -np.log(np.sqrt(2*np.pi)*sigma[0,:]) -0.5*np.power(np.divide(
                    (x-mu[0,:]),sigma[0,:]),2),axis=1)
    
    return CLASSES[np.argmax([critere_1, critere_2], axis=0)]
```

```python
def predict_label(x):
    # Cette fonction est utlisée pour l'affichage de la frontiere
    CLASSES = np.array([-1,1])
    
    critere_2 = logp2 + np.sum(
        -np.log(np.sqrt(2*np.pi)*sigma[1,:]) -0.5*np.power(np.divide(
                    (x-mu[1,:]),sigma[1,:]),2),axis=1)
    
    critere_1 = logp2 + np.sum(
        -np.log(np.sqrt(2*np.pi)*sigma[0,:]) -0.5*np.power(np.divide(
                    (x-mu[0,:]),sigma[0,:]),2),axis=1)
    
    return CLASSES[np.argmax([critere_1, critere_2], axis=0)]  
```

## 2 Descente de gradient
L’algorithme de descente de gradient est un algorithme itératif qui permet d’approcher le minimum
d’une fonction convexe f donnée. Il consiste à calculer à chaque étape t le gradient de la fonction
convexe au point courant $ x^t $
et de mettre-à-jour le point courant en se déplaçant dans le sens opposé
au gradient :$ x
^{t+1} = x
^t − \epsilon∗ ∇f(x^
t
).$

### Q 2.1 
Coder une fonction optimize(fonc,dfonc,xinit,eps,max_iter) qui implémente l’algorithme
du gradient avec en paramètre : fonc la fonction à optimiser, dfonc le gradient de cette fonction, xinit
le point initial, eps le pas de gradient, max_iter le nombre d’itérations. Cette fonction doit rendre un
triplet (x_histo,f_histo,grad_histo), respectivement la liste des points $x^t
,f(x^
t
)$ et $∇f(x^
t
)$. Pour
cela vous pouvez utiliser une liste pour stocker au fur et à mesure les points, puis transformer chaque
liste en un tableau np.array.

```python
def optimize(fonc, dfonc, xinit, eps, max_iter, **kwargs):
    """
    Descente de gradient,
    fonc: fonction a optimiser
    dfonc: gradient de la fonction,
    eps: pas de l'algorithme,
    max_iter: nombre max d'itérations
    retourne la liste ( x, f(x), df(x)) pour chaque itération
    """
    return_list = []
    if "data" in kwargs.keys():
        return_list.append( [xinit, fonc(xinit, kwargs['data'], kwargs['label']), dfonc(xinit, kwargs['data'], kwargs['label'])])
        for i in range(0,max_iter):
            xpred, fpred, dfpred= return_list[-1]
            x = xpred - eps*dfpred
            return_list.append([x, fonc(x, kwargs['data'], kwargs['label']), dfonc(x,kwargs['data'],kwargs['label'])])
    else:
        return_list.append( [xinit, fonc(xinit), dfonc(xinit)])
        for i in range(0,max_iter):
            xpred, fpred, dfpred= return_list[-1]
            x = xpred - eps*dfpred
            return_list.append([x, fonc(x), dfonc(x)])
    
    return np.array(return_list)
    
```

 ## 3 Applications : Classification plug-in et Régression logistique
 
On suppose dans la suite que le problème de classification est binaire à valeur dans {−1, +1}. Nous
noterons $η(x) = P(y = 1|x) $et$ X = \{x^
i
, y^i\}_{i=1}^N
$l’ensemble d’apprentissage. 

Dans cette partie, vous
allez implémenter deux approches de classification binaire, la classification par plug-in en utilisant une
régression linéaire, et la régression logistique. Vous utiliserez la descente de gradient pour optimiser les
paramètres du classifieur.

Asstuce d’implémentation : comme dans le cas de la régression linéaire, nous utiliserons des fonctions
linéaires dont le résultat peut s’exprimer par produit scalaire. Afin d’incorporer le biais $w_0$ dans le
produit scalaire et de soulager les notations et les lignes de codes, il est conseillé d’ajouter à la matrice
des données une colonne de 1 en première colonne par exemple. La transformation par $f_w$ correspond
alors exactement à un produit scalaire : $f_w(x) =< w,(1, x_1, x_2, . . . , x_d >$

### Classification plug-in
Dans ce contexte, le problème de classification est vu comme un problème
de régression aux moindres carrés : il s’agit de trouver la fonction $f^∗$ qui minimise $E[(f^
∗
(x) − y)^
2
]$
(soit trouver f
qui approxime au mieux 2η(x) − 1). Nous supposerons f linéaire, paramétrée par un vecteur de poids $w : f_w(x) = \sum_{i=1}^{d} wi*xi + w0.$

• Inférence : le signe de $f_w(x)$ conduit à la classification de x : +1 si positif, −1 sinon.

• Coût à optimiser :
$ L(w, X) = \sum_{i=1}^{N}(f_w(x) − y)^2 =\sum_{i=1}^{N}(w0 +
\sum_{j=1}^{d} w_{j}*x_{j}^{i} − y^{i})^2 $


### Régression logistique 
Ce modèle considère que la probabilité η(x) peut être modélisé par $ σ (fw(x)) =
1+e^{−f_w(x)}$
, $f_w$ fonction linéaire ;

ainsi, $ log( \frac {
η(x)}{
1−η(x)})$
est approximée par $f_w(x)$).


Par ailleurs, $ P(y = −1|x) = 1 − η(x) = σ(−f_w(x))$. 
Comme dans le cas du classifieur bayésien naïf,
on utilise le maximum de vraisemblance pour trouver les paramètres optimaux (ou plus exactement
minimiser la neg-log vraisemblance).

• Inférence : classer en +1 si $σ (f_w(x)) > 0.5$, en −1 sinon ; soit classifier selon le signe de $f_w(x)$.

• Coût à optimiser : $L(w, X) = \sum_
{i=1}^{N} log(
1 + e^{
(−y^
i*f_w(x^
i
)}) = \sum_{ i=1}^{N} log( 1 + e^{(−y^
i(w_0+\sum_{j=1}^{d} w_jx^i_j )})$

```python
def mse(w, data, label):
    """
    Retourne  la mse pour la classification plug-in
    w, vecteur w
    label, vecteur des labels des classes;
    data, matrice des donnees
    """
    data = np.hstack((np.ones((data.shape[0],1)),data))
    temp_tab = np.power(data.dot(w) - label.T,2)
    return np.sum(temp_tab)
```

```python
def grad_mse(w, data, label):
    """
    Retourne  le gradient de la mse pour la classification plug-in
    w, vecteur w
    label, vecteur des labels des classes;
    data, matrice des donnees
    """
    data = np.hstack((np.ones((data.shape[0],1)),data))
    temp_tab = (data.dot(w) -label.T).dot(data)
    return  temp_tab.reshape(-1)
```

### Q 3.3
Implémenter les fonctions reglog(w,data,label) et grad_reglog(w,data,label) qui calculent
le coût en w et le gradient du coût pour la régression logistique.

```python
def reglog(w, data, label):
    """
    Retourne  la fonction de cout pour la classification regression logistique
    w, vecteur w
    label, vecteur des labels des classes;
    data, matrice des donnees
    """
    data = np.hstack((np.ones((data.shape[0],1)),data))
    temp_tab = np.log(1+np.exp(np.multiply(label.T,data.dot(w))))
    return np.sum(temp_tab)
```

```python
def grad_reglog(w, data, label):
    """
    Retourne  le gradient fonction de cout pour la classification regression logistique
    w, vecteur w
    label, vecteur des labels des classes;
    data, matrice des donnees
    """
    data = np.hstack((np.ones((data.shape[0],1)),data))
    label = label.reshape(-1,1)
    temp_tab = np.multiply(np.multiply(data, label).T,np.exp(np.multiply(label.T,data.dot(w))))/(1+np.exp(np.multiply(label.T,data.dot(w))))
    return np.sum(temp_tab, axis=1).reshape(-1)
```

```python
def fit_plugin( xdata, ydata, pas, max_iter):
    """
    Fit le modele de classification plug-in
    """
    dic = {'data':xdata,'label':ydata}
    w = np.zeros((xdata.shape[1]+1,1)).reshape(-1)
    temp = optimize(mse, grad_mse, w, pas, max_iter, **dic)
    return temp[-1, 0]
```

```python
def predict_label_plugin_front(xdata):
    """
    Fonction pour afficher la frontiere
    """
    xdata = np.hstack((np.ones((xdata.shape[0],1)),xdata))
    temp_tab = xdata.dot(w)
    return np.sign(temp_tab)
```

```python
def fit_reglog(xdata, ydata, pas, max_iter):
    """
    Fit le modele de classification regression logistique
    """
    dic = {'data':xdata,'label':ydata}
    w = np.zeros((xdata.shape[1]+1,1)).reshape(-1)
    temp = optimize(reglog, grad_reglog, w, pas, max_iter, **dic)
    return temp[-1, 0]
```

```python
def predict_reglog(w,xdata):
    """
    Prediction des labels pour la regression logistique
    """
    xdata = np.hstack((np.ones((xdata.shape[0],1)),xdata))
    temp_tab = xdata.dot(w)
    return -np.sign(temp_tab)
```

```python
def predict_reglog_front(xdata):
    """
    Fonction pour afficher la frontiere
    """
    xdata = np.hstack((np.ones((xdata.shape[0],1)),xdata))
    temp_tab = xdata.dot(w)
    return -np.sign(temp_tab)
```

```python
def predict_label_plugin(w, xdata):
    """
    Prediction des labels pour la classification plug-in
    """
    xdata = np.hstack((np.ones((xdata.shape[0],1)),xdata))
    temp_tab = xdata.dot(w)
    return np.sign(temp_tab)
```
